# Notación asintótica

**_Notación Asintótica:_** Trata acerca del comportamiento de funciones en el límite, es decir, en valores lo suficientemente grandes de su parámetro.

Los argumentos basados en *notación asintótica* pueden no llegar a tener un valor práctico en la vida real, pero su estudio es de vital importancia para el entendimiento de los algoritmos y su funcionamiento.

## 3.1 Notación para el "*orden de*": 

### Cota Superior -- Notación $\mathcal{O}(n)$ (*"O grande"* o *"Big Oh"* (en inglés))

Dada una función $f$, se desea estudiar aquellas funciones $t$ que a lo sumo crecen tan deprisa como $f$. Al conjunto de tales funciones se le llama *cota superior de $f$* y se denomina $\mathcal{O}(f(n))$

sea $t:\mathbb{N} \rightarrow \mathbb{R}^{\geq 0}$ una función arbitraria de los números naturales en los números reales no negativos, como por ejemplo, $t(n)=27n^2+\frac{355}{113}n+12$. $n$ representa el tamaño del ejemplar sobre el cual se aplicará el algoritmo dado, y $t(n)$ es la cantidad de un recurso dado que se invierte en ese ejemplar (tiempo, espacio, por ejemplo).

- $t(n)$ puede depender más de la implementación que del algoritmo en ciertos casos.

Ahora consideremos la función $f:\mathbb{N} \rightarrow \mathbb{R}^{\geq 0}$ tal como $f(n)=n^2$. Diremos que $t(n)$ está en el *orden de* $f(n)$ si $t(n)$ está acotada superiormente por un múltiplo real positivo de $f(n) \forall n$ suficientemente grande.

- Matemáticamente esto significa que existe una constante real y positiva $c$ y un entero *umbral* $n_0$ tal que $t(n) \leq cf(n)$, siempre que $n \geq n_0$

Retomando las funciones definidas anteriormente, es claro que tanto $n \leq n^2$ como $1\leq n^2$, siempre que $n \geq 1$, entonces:

$$t(n)=27n^2+\frac{355}{113}n+12$$
$$\leq 27n^2+\frac{355}{113}n^2+12n^2$$
$$=\frac{4762}{113}n^2=\frac{4762}{113}f(n)$$


Tomando $c=\frac{4762}{113}$ (o cualquier valor mayor) y $n_0=1$, se concluye entonces que $t(n)$ es del orden de $f(n)$ ya que $t(n) \leq cf(n)$ siempre que $n \geq n_0$.

De esta manera, si la implementación de un algoritmo requiere en el peor caso $27n^2+\frac{355}{113}n+12$ microsegundos para resolver un caso de tamaño $n$, se puede simplificar diciendo que el tiempo está en el orden de $n^2$.

- Cuando se habla en términos del *orden de* no se establecen las dimensiones, pues no tiene sentido indicar que serían microsegundos o años, ya que estas cantidades solo se diferencian por un factor constante.


- El *"orden de"* caracteriza no solo el requerido por una implementación particular del algoritmo, sino también (por el *principio de invarianza*) el tiempo requerido por cualquier implementación.

Por conveniencia, se dispone de un símbolo matemático para representar el *orden de* y se define a continuación.

> Sea $f:\mathbb{N} \rightarrow \mathbb{R}^{\geq 0}$ una función arbitraria de los números naturales en los reales no negativos. Se indicará mediante $\mathcal{O}(f(n))$ que se lee *"O de f(n)"* e indica el conjunto de todas las funciones $f:\mathbb{N} \rightarrow \mathbb{R}^{\geq 0}$ tales que $f(n) \leq cf(n)$ para todo $n\geq n_0$ para una constante positiva real $c$ y para un umbral entero $n_0$. 

En otras palabras:

$$\mathcal{O}(f(n))=\{t:\mathbb{N} \rightarrow \mathbb{R}^{\geq 0}|(\exists c \in \mathbb{R}^+)(\overset{\infty}{\forall} n \in \mathbb{N})[t(n)\leq cf(n)]\}$$

***Regla del máximo:*** Sean $f,g:\mathbb{N} \rightarrow \mathbb{R}^{\geq 0}$ dos funciones arbitrarias de los números naturales en los reales no negativos. Entonces, $\mathcal{O}(f(n)+g(n))=\mathcal{O}(máx(f(n),g(n)))$.

Como ejemplo, asumamos que se tiene un algoritmo que se realiza en tres pasos: inicialización, procesamiento y finalización. Supongamos también que estos pasos requieren un tiempo de $\mathcal{O}(n^2), \mathcal{O}(n^3)$ y $\mathcal{O}(nlog(n))$ respectivamente. El algoritmo completo requiere un tiempo de $\mathcal{O}(n^2+n^3+nlog(n))$. Aplicando la regla del máximo tenemos:

$$\mathcal{O}(n^2+n^3+nlog(n))=\mathcal{O}(máx(n^2,n^3,nlog(n)))$$
$$=\mathcal{O}(n^3)$$

Aunque el tiempo requerido por el algoritmo es la suma de los tiempos requeridos por cada una de las partes por separado, el tiempo es del orden del tiempo de la parte que consuma más tiempo.

Si se tiene una función complicada tal como $t(n)=12n^3log(n)-5n^2+log^2n+36$, y si $f(n)=n^3log(n)$ es el término más significativo de $t(n)$ descartando el coeficiente, entonces $\mathcal{O}(t(n)) = \mathcal{O}(f(n))$. 

- Esto permite una simplificación muy poderosa de la notación asintótica.


- Se pueden despreciar los términos de orden inferior porque son despreciables en comparación con el término de orden superior para $n$ suficientemente grande.

- No se debe demostrar que $\mathcal{O}(t(n)) = \mathcal{O}(f(n))$ razonando de la siguiente forma:

$$\mathcal{O}(t(n)) = \mathcal{O}(máx(12n^3log(n),-5n^2,log^2n+36))$$

$$\mathcal{O}(12n^3log(n))=\mathcal{O}(n^3log(n))$$

- De esta forma no se está empleando adecuadamente la regla del máximo, ya que hay una función negativa.

Veamos el siguiente razonamiento:

$$\mathcal{O}(t(n)) = \mathcal{O}(11n^3log(n)+n^3log(n),-5n^2 + log^2n + 36))$$

$$ = \mathcal{O}(máx(11n^3log(n), n^3log(n), -5n^2 + log^2n + 36))$$

$$=\mathcal{O}(11n^3log(n))=\mathcal{O}(n^3log(n))$$

- Obsérvese que $n^3log(n) -5n^2$ es negativo  y $36$ es mayor que $11n^3log(n)$ para valores de $n$ pequeños, sin embargo, esto no es así para valores de $n$ suficientemente grandes.

- Otra observación útil es que normalmente no es necesario especificar la base del logaritmo dentro de la notación asintótica, ya que $log_a(n) = log_a(b) \times log_b(n)$ para todos los reales positivos $a$, $b$ y $n$ tales que ni $a$ ni $b$ es igual a $1$. 


- Lo importante es que $log_a(b)$ es una constante positiva, cuando $a$ y $b$ son constantes mayores que $1$. Por lo que $log_a(n)$ y $log_b(n)$ solo difieren en una constante multiplicativa.


- Resulta fácil ver entonces que $\mathcal{O}(log_a(n)) = \mathcal{O}(log_b(n))$, lo que se simplifica como $\mathcal{O}(log(n))$

- La base del logaritmo no se puede ignorar cuando es $<1.0$, cuando no es una constante, $\mathcal{O}(log_n(n)) \neq \mathcal{O}(log(n)$, o cuando el logaritmo se encuentra en el exponente, tal como en $\mathcal{O}(2^{ln(n)}) \neq \mathcal{O}(2^{log(n)})$

***Regla del límite:*** se emplea para demostrar que algunas funciones están en el orden de otras y para demostrar lo contrario.

Dadas las funciones arbitrarias $f$ y $g:\mathbb{N} \rightarrow \mathbb{R}^{\geq 0}:$

1. si $\lim_{n \to \infty} \frac{f(n)}{g(n)} \in \mathbb{R}^+$ entonces $f(n) \in \mathcal{O}(g(n))$ y $g(n) \in \mathcal{O}(f(n))$

2. si $\lim_{n \to \infty} \frac{f(n)}{g(n)} = 0$ entonces $f(n) \in \mathcal{O}(g(n))$ pero $g(n) \not\in \mathcal{O}(f(n))$, y

3. si $\lim_{n \to \infty} \frac{f(n)}{g(n)} = +\infty$ entonces $f(n) \not\in \mathcal{O}(g(n))$ pero $g(n) \in \mathcal{O}(f(n))$


Existen otras dos notaciónes asintóticas

- ***Cota inferior o Notación $\Omega$ (Omega):*** Funciones $t$ que a lo sumo crecen tan lentamente como $f$.

> $g(n) = \Omega (f(n))$ sii $|g(n)/f(n)|$ es limitado desde abajo cuando $n \rightarrow \infty$ 



$$\mathcal{O}(f(n))=\{t:\mathbb{N} \rightarrow \mathbb{R}^{\geq 0}|(\exists c \in \mathbb{R}^+)(\overset{\infty}{\forall} n \in \mathbb{N})[t(n)\leq cf(n)]\}$$

- ***Orden exacto o Notación $\Theta$ (Theta):*** Conjunto de funciones que crecen asintóticamente de la misma forma (dentro de un factor constante).


> $g(n) = \Theta (f(n))$ sii $g(n)= \mathcal{O}(f(n))$ y $g(n) = \Omega(f(n))$

## 3.2 Operaciones sobre notación asintótica

Para simplificar los cálculos, se puede manipular la notación asintótica empleando operadores aritméticos.

***Por ejemplo:***

$\mathcal{O}(f(n))+ \mathcal{O}(g(n))$ representa el conjunto de operaciones obtenidas sumando punto a punto toda función de $\mathcal{O}(f(n))$ a cualquier función de $\mathcal{O}(g(n))$


- Este conjunto representa el orden del tiempo requerido por un algoritmo compuesto por una fase que requiere un tiempo de orden $f(n)$ seguido por una segunda fase que requiere un tiempo del orden de $g(n)$.


- Se puede demostrar que $\mathcal{O}(f(n))+ \mathcal{O}(g(n))$ es idéntico a $\mathcal{O}(f(n) + g(n))$ , y por la regla del máximo, esto también es igual a $\mathcal{O}(máx(f(n),g(n)))$ 


Para finalizar, en la siguiente tabla se muestra una serie de funciones de orden de magnitud más comunes encontradas en el estudio de los algoritmos


| $f$ $($ $n$ $)$      | Nombre    |
|:--------------------:|:---------:|
|1                     |Constante  |
|$log$ $($ $n$ $)$     |Logaritmica|
|$n$                   |Lineal     |
|$n$ $log$ $($ $n$ $)$ |Log-lineal |
|$n^2$                 |Cuadrática |
|$n^3$                 |Cúbica     |
|$2^n$                 |Exponencial|

```python
import numpy as np
import matplotlib.pyplot as plt

def grafica(cte,lgn,n1,nlgn,n2,n3,dosn,n):
    x = np.arange(1, n)
    #plt.figure(figsize=(10, 8))
    plt.semilogy(x,cte, label = "$Constante$")
    plt.semilogy(x,lgn, label = "$Logarítmica$")
    plt.semilogy(x,n1, label = "$Lineal$")
    plt.semilogy(x,nlgn, label = "$Log-lineal$")
    plt.semilogy(x,n2, label = "$Cuadrática$")
    plt.semilogy(x,n3, label = "$Cúbica$")
    plt.semilogy(x,dosn, label = "$Exponencial$")
    plt.legend()
    #plt.title("Algoritmo vs hardware")
    plt.xlabel("Tamaño $n$ del problema")
    plt.ylabel("log(y)")  # Colocamos la etiqueta en el eje y
    plt.axhline(y=0, color='k')
    plt.xlim(1, n-1) 
    plt.grid(True)
    plt.show()

cte,lgn,n1,nlgn,n2,n3,dosn,n = [],[],[],[],[],[],[],15

c1 = 1

for i in range(1,n):
    cte.append(10*n)
    lgn.append(c1*np.log(i))
    n1.append(c1*i)
    nlgn.append(c1*i*np.log(i))
    n2.append(c1*i**2)
    n3.append(c1*i**3)
    dosn.append(c1*2**i)    
grafica(cte,lgn,n1,nlgn,n2,n3,dosn,n)
```
